```python
# Importation des bibliothèques utiles
import pandas as pd 
import numpy as np
from math import isnan
import PyPaillier.phe.paillier as he
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
# Génération d'un couple (clé publique, clé privée) côté client
n_length = 100
pubkey, prikey = he.generate_paillier_keypair(n_length=n_length)

# Le client envoie deux nombres chiffrées au serveur, à savoir 2 et 3
encryptedNumberA = pubkey.encrypt(2)
encryptedNumberB = pubkey.encrypt(3)

# Le serveur peut ajouter ces deux nombres chiffrés sans les connaître
encryptedNumberC = encryptedNumberA.__add__(encryptedNumberB)

# Voici la seule information (la somme chiffrée) que le serveur possède
print("Information connue du serveur : {0}".format(encryptedNumberC.ciphertext()))

# Le client est le seul à pouvoir déchiffrer cette somme !
print("Somme déchiffrée par le client : {0}".format(prikey.decrypt(encryptedNumberC)))
```

```python
# Chargement du dataset
train_data = pd.read_csv('data/data.csv', sep=',')
train_data.set_index('id', inplace=True, drop=True)

print("Dimensions: {0}".format(train_data.shape))
train_data.head()
```

```python
# Fonctions utiles au chiffrement du dataset
M = [[254125422,4,3],[1,2,1],[8,9,1]]

def encrypt_vector(public_key, x):
    return [public_key.encrypt(i,0.05) for i in x]

def encrypt_matrix(public_key, X):
    return [encrypt_vector(public_key, x) for x in X]

def encrypted_vector_ciphertext(encrypted_x):
    return [i.ciphertext() for i in encrypted_x]

def encrypted_matrix_ciphertext(encrypted_X):
    return [encrypted_vector_ciphertext(encrypted_x) for encrypted_x in encrypted_X]


encrypted_matrix_ciphertext(encrypt_matrix(pubkey,M))
```

```python
# Fonctions utiles à l'analyse du dataset

# Fonction pour le scaling des données 
def scaling(X):
    scaler = MinMaxScaler()
    scaler.fit(X)
    return scaler.transform(X)

# Fonction de cross-validation
def compute_score(clf, X, y):
    xval = cross_val_score(clf, X, y, cv=5)
    return np.mean(xval)
```

```python
# Entraînement et validation du modèle 0 : Quick & Dirty (données en clair)
def parse_model0(X):
    # Définition de la target 
    target = X.diagnosis
    
    # Suppression des features inutiles
    to_del = ['diagnosis', 'Unnamed: 32']
    for col in to_del : del X[col]
        
    # Scaling des données
    X = scaling(X)

    return X, target

X,y = parse_model0(train_data.copy())
lr = LogisticRegression()
compute_score(lr,X,y)
```

```python
# Nous définissons ci-dessous des fonctions permettant de transformer une matrice de floats en matrice d'entiers
# Ceci est utile dans notre cas, car nous ne pouvons pas appliquer un chiffremement homomorphe à des valeurs décimales

# Fonction retournant la puissance de 10 par laquelle multiplier les composantes d'un vecteur de floats U
def float_multiplier_power(U):    
    max_decimal_part = 0
    for u in U :
        if '.' in str(u):
            u_decimal_part_size = len(str(u).split('.')[1])
        else :
            u_decimal_part_size = 0
        if (u_decimal_part_size > max_decimal_part) :
            max_decimal_part = u_decimal_part_size
    return max_decimal_part

# Fonction multipliant toutes les composantes d'un vecteur de floats U par la puissance de 10 définie par float_multiplier
def vector_multiply(U):
    max_decimal = float_multiplier_power(U)
    for i in range(len(U)):
        if '.' in str(U[i]):
            u_decimal_part_size = len(str(U[i]).split('.')[1])
        else :
            u_decimal_part_size = 0
        u_string = (str(U[i])+"0"*(max_decimal-u_decimal_part_size)).replace('.','')
        compteur_zero=0
        for j in range(len(u_string)):
            if u_string[j] == '0':
                compteur_zero=compteur_zero+1
            else:
                break
        U[i]=int(u_string[compteur_zero:])
    return U

# Fonction multipliant toutes les composantes de tous les vecteurs d'une matrice X par le nombre adéquat
def floatMatrix_to_intMatrix(X):
    for i in range(len(X)):
        # Les lignes exclues ci-dessous causaient un problème lors de la conversion de la matrice floatMatrix_to_intMatrix
        if (i not in [38,101,140,166,174,175,178,192,241,277,311,313,314,376,391,412,473,538,550,557,561,568]):
            vector_multiply(X[i])
    return X
```

```python
# Entraînement et validation du modèle 1 (données chiffrées)
def parse_model1(X):
    # Définition de la target 
    target = X.diagnosis
    
    # Suppression des features inutiles
    to_del = ['diagnosis', 'Unnamed: 32']
    for col in to_del : del X[col]
    
    # Scaling des données
    X = scaling(X)

    # Transformation de la matrice de floats en matrice d'entiers
    np.set_printoptions(precision=6)
    np.set_printoptions(suppress=True)
    X = np.array(X, float)
    X = np.array(floatMatrix_to_intMatrix(X),int)

    # Chiffrement homomorphe des données
    X = encrypted_matrix_ciphertext(encrypt_matrix(pubkey,X))
    
    return X, target

X,y = parse_model1(train_data.copy())
lr = LogisticRegression()
compute_score(lr,X,y)
```
